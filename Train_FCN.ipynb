```python
import random
import numpy as np
import imageio

%matplotlib inline
import matplotlib.pyplot as plt

import tensorflow as tf
```

Загрузим датасет(одно фото и карта сегментации)

```python
img = imageio.imread('6.11_me_256.png')
img = img.astype(np.float32) / 255
_=plt.imshow(img)
```

```python
gt = imageio.imread('6.11_me_gt_256.png')
gt = gt.astype(np.float32) / 255.
gt = np.mean(gt, axis=2)[..., None]
_=plt.imshow(gt[..., 0])
```

Подготовим обучающий датасет из 2-х картинок

```python
train_x = np.stack((img, img[:, ::-1,:]))
train_y = np.stack((gt, gt[:, ::-1, :]))

plt.imshow(train_x[0]), plt.show()
plt.imshow(train_y[0])[..., 0], plt.show()
plt.imshow(train_x[1]), plt.show()
plt.imshow(train_y[1])[..., 0], plt.show()
```
