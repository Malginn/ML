```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pickle

from sklearn.cluster import DBSCAN
```

```python
with open('clustering.pkl', 'rb') as f:
    data_clustering = pickle.load(f)

data_clustering     #Y - не нужен

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])
```

```python
db = DBSCAN(eps=0.6, min_samples=10).fit(X)

# core_samples_mask = np.zeros_like(db.labels_, dtype=bool)   #создаем массив нулевых значений
# core_samples_mask[db.core_sample_indices_] = True   #для опорных значений делаем true
```

```python
label= db.labels_

plt.scatter(X[:, 0], X[:, 1], s=40, c=label, marker='o')
plt.show()
```

```python
db = DBSCAN(eps=0.6, min_samples=5).fit(X)

label= db.labels_

plt.scatter(X[:, 0], X[:, 1], s=40, c=label, marker='o')
plt.show()
```
