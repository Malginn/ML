```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pickle
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, adjusted_mutual_info_score
```

```python
with open('clustering.pkl', 'rb') as f:
    data_clustering = pickle.load(f)

data_clustering     #Y - не нужен

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])
```

```python
model = KMeans(n_clusters=2, random_state=0).fit(X)

score = silhouette_score(X, model.labels_)
print('Качество кластеризации по метрике силуэтта %.3f' % score)

plt.scatter(X[:, 0], X[:, 1], c=model.labels_)
plt.show()
```

```python
model3 = KMeans(n_clusters=3, random_state=0).fit(X)

score3 = silhouette_score(X, model3.labels_)
print('Качество кластеризации по метрике силуэтта %.3f' % score3)

plt.scatter(X[:, 0], X[:, 1], c=model3.labels_)
plt.show()
```

Нужно меньше смотреть глазами и верить метрикам

```python
model3 = KMeans(n_clusters=3, random_state=0).fit(X)

score3 = adjusted_mutual_info_score(Y, model3.labels_, average_method='arithmetic')
print('Качество кластеризации по метрике AMI %.3f' % score3)

plt.scatter(X[:, 0], X[:, 1], c=model3.labels_)
plt.show()
```
